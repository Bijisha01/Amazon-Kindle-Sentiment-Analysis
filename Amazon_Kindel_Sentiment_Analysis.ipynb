```python
# 3 classes - Negative, Neutral, Positive

def rating_to_sentiment(final):
    def map_rating_to_sentiment(rating):
        if rating in [1, 2]:
            return 'Negative'
        elif rating == 3:
            return 'Neutral'
        else:
            return 'Positive'

    # Apply the mapping function to the 'overall' column
    final['ratings_sentiment'] = final['overall'].apply(map_rating_to_sentiment)

    return final
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

# Download necessary NLTK data
nltk.download('vader_lexicon')
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab') # This line downloads the missing 'punkt_tab' data


# Create a SentimentIntensityAnalyzer object
sia = SentimentIntensityAnalyzer()

def preprocess_and_score(text):
    # Tokenization
    tokens = nltk.word_tokenize(text)

    # Remove stop words
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]

    # Stemming
    stemmer = PorterStemmer()
    tokens = [stemmer.stem(token) for token in tokens]

    # Join tokens back into text
    preprocessed_text = ' '.join(tokens)

    # Calculate sentiment score
    sentiment = sia.polarity_scores(preprocessed_text)
    return sentiment['compound']

# Apply the function to the 'reviewText' column in the final DataFrame
final['review_sentiment_score'] = final['reviewText'].apply(preprocess_and_score)
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')

sia = SentimentIntensityAnalyzer()

def preprocess_and_score(text):
    if not isinstance(text, str):
        return 0  # Return neutral sentiment for missing or invalid text
    # Tokenization
    tokens = word_tokenize(text)
    # Remove stop words
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word.lower() not in stop_words]
    # Reconstruct cleaned text
    cleaned_text = ' '.join(tokens)
    # Sentiment scoring
    sentiment = sia.polarity_scores(cleaned_text)
    return sentiment['compound']

# Apply the function to the 'reviewText' column in the final DataFrame
final['review_sentiment_score'] = final['reviewText'].apply(preprocess_and_score)
```

# Modeling

```python
#Random Forest
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Step 1: Define sentiment target based on Ratings
final['sentiment'] = final['Ratings'].apply(lambda x: 1 if x > 3 else 0)  # Positive sentiment (1), negative (0)

# Step 2: Preprocess reviewText using TF-IDF
vectorizer = TfidfVectorizer(max_features=500)  # Adjust max_features as needed
tfidf_features = vectorizer.fit_transform(final['reviewText'].fillna("")).toarray()

# Combine features and calculate sentiment score per book
X_sentiment = tfidf_features
y_sentiment = final['sentiment']

# Split data for sentiment model
X_train, X_test, y_train, y_test = train_test_split(X_sentiment, y_sentiment, test_size=0.2, random_state=42)

# Step 3: Train the Sentiment Classifier
sentiment_model = RandomForestClassifier(random_state=42, n_estimators=100)
sentiment_model.fit(X_train, y_train)

# Predict sentiment probabilities for the entire dataset
final['sentiment_score'] = sentiment_model.predict_proba(X_sentiment)[:, 1]  # Probability of positive sentiment

# Step 4: Use Sentiment Scores and Other Features to Predict Bestseller Status
bestseller_features = ['sentiment_score', 'Ratings', 'isKindleUnlimited', 'isEditorsPick']  # Example features
X_bestseller = final[bestseller_features]
y_bestseller = final['isBestSeller'].astype(int)  # Ensure target is numeric

# Split data for bestseller prediction
X_train_b, X_test_b, y_train_b, y_test_b = train_test_split(X_bestseller, y_bestseller, test_size=0.2, random_state=42)

# Step 5: Train the Bestseller Classifier
bestseller_model = RandomForestClassifier(random_state=42, n_estimators=100)
bestseller_model.fit(X_train_b, y_train_b)

# Step 6: Evaluate the Bestseller Model
y_pred_b = bestseller_model.predict(X_test_b)
y_pred_proba_b = bestseller_model.predict_proba(X_test_b)[:, 1]

conf_matrix_b = confusion_matrix(y_test_b, y_pred_b)
sns.heatmap(conf_matrix_b, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix - Bestseller Prediction')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

print("Accuracy (Bestseller):", accuracy_score(y_test_b, y_pred_b))
print("Classification Report (Bestseller):\n", classification_report(y_test_b, y_pred_b))

roc_auc_b = roc_auc_score(y_test_b, y_pred_proba_b)
print("ROC AUC (Bestseller):", roc_auc_b)

# Plot ROC Curve for Bestseller
fpr_b, tpr_b, _ = roc_curve(y_test_b, y_pred_proba_b)
plt.plot(fpr_b, tpr_b, label=f'Random Forest - Bestseller (AUC = {roc_auc_b:.2f})')
plt.plot([0, 1], [0, 1], linestyle='--', label='Random')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - Bestseller Prediction')
plt.legend()
plt.show()
```

#Predicting Bestseller Status Based on Book Name

```python
def predict_bestseller(book_name, final, sentiment_model, bestseller_model, vectorizer):
    """
    Predict whether a given book is a bestseller based on its name and associated data.

    Args:
        book_name (str): Name of the book to predict.
        final (pd.DataFrame): DataFrame containing book information.
        sentiment_model: Pre-trained sentiment analysis model.
        bestseller_model: Pre-trained bestseller prediction model.
        vectorizer: Pre-trained TF-IDF vectorizer for `reviewText`.

    Returns:
        str: Prediction result ("Bestseller" or "Not a Bestseller").
    """
    # Step 1: Find the book in the DataFrame
    book_data = final[final['title'].str.contains(book_name, case=False, na=False)]

    if book_data.empty:
        return f"Book '{book_name}' not found in the dataset."

    # Step 2: Process the reviewText for sentiment analysis
    review_texts = book_data['reviewText'].fillna("")
    tfidf_features = vectorizer.transform(review_texts).toarray()

    # Predict sentiment scores for the book
    sentiment_scores = sentiment_model.predict_proba(tfidf_features)[:, 1]
    avg_sentiment_score = sentiment_scores.mean()  # Average sentiment score for the book

    # Step 3: Prepare features for bestseller prediction
    book_features = book_data[['Ratings', 'isKindleUnlimited', 'isEditorsPick']].iloc[0].values
    features_for_bestseller = [avg_sentiment_score] + book_features.tolist()

    # Step 4: Predict bestseller status
    bestseller_prediction = bestseller_model.predict([features_for_bestseller])[0]

    # Step 5: Return the result
    if bestseller_prediction == 1:
        return f"The book '{book_name}' is predicted to be a Bestseller."
    else:
        return f"The book '{book_name}' is predicted to NOT be a Bestseller."


```
